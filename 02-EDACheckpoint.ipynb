#### EDA: Distribution Analysis

This section  aims to understand the basic structure and behavior of each relevant variable in the dataset before analyzing relationships between variables

##### Basic dataset characteristics

```python
import pandas as pd
import numpy as np
from pathlib import Path

data_path = Path('data/00-raw/Students Social Media Addiction.csv')
if not data_path.exists():
    data_path = Path('Group029_WI26/data/00-raw/Students Social Media Addiction.csv')

ssma = pd.read_csv(data_path)

print("Rows, cols:", ssma.shape)
print("\nMissing values per column:\n", ssma.isna().sum().sort_values(ascending=False))
print("\nUnique counts:\n", ssma.nunique().sort_values(ascending=False))
```

Interpretation
- Dataset contains 705 students and all columns are complete

- Key grouping variables (Gender, Academic_Level) have limited categories and are suitable for stratified analyses

- Some fields (Country) have many levels, so later visualizations can use top-N aggregation

##### Numeric distributions

Variables to treat as numeric:
- Addicted_Score (main predictor)
- Mental_Health_Score (main outcome)
- Avg_Daily_Usage_Hours
- Sleep_Hours_Per_Night
- Conflicts_Over_Social_Media
- Age

```python
num_cols = [
    "Age","Avg_Daily_Usage_Hours","Sleep_Hours_Per_Night",
    "Mental_Health_Score","Conflicts_Over_Social_Media","Addicted_Score"
]

summary = ssma[num_cols].describe(percentiles=[.05,.25,.5,.75,.95]).T
summary["skew"] = ssma[num_cols].skew(numeric_only=True)
summary["kurtosis"] = ssma[num_cols].kurtosis(numeric_only=True)
summary
```

Interpretation:
- Age ranges from 18 to 24, with a mean of 20.66 and low standard deviation showing the sample is concentrated around typical college-aged students
- Average daily social media usage ranges from 1.5 to 8.5 hours, with a mean of 4.92 hours suggesting moderate variability in usage intensity
- Sleep hours per night average 6.87 hours (range 3.8–9.6), indicating some students experience relatively low sleep levels
- Mental health scores range from 4 to 9, with a mean of 6.23 and median of 6, suggesting most students report moderate mental health levels with reasonable spread
- Social media addiction scores range from 2 to 9, with a mean of 6.44, indicating that many students report moderately high levels of social media addiction.
- Skewness values for all variables are close to zero, suggesting approximately symmetric distributions, and kurtosis values are slightly negative, indicating light-tailed distributions without extreme outliers

##### Histograms

```python
import numpy as np
import matplotlib.pyplot as plt

plots = [
    ("Addicted_Score", "Addicted Score", np.arange(1.5, 10.5, 1)),            # bins centered on integers 2..10
    ("Mental_Health_Score", "Mental Health Score", np.arange(0.5, 10.5, 1)),  # bins centered on integers 1..10
    ("Avg_Daily_Usage_Hours", "Avg Daily Usage (Hours)", 10),                 # continuous
    ("Sleep_Hours_Per_Night", "Sleep Hours Per Night", 10),                   # continuous
    ("Conflicts_Over_Social_Media", "Conflicts Over Social Media", np.arange(-0.5, 6.5, 1)),  # 0..5
]

fig, axes = plt.subplots(2, 3, figsize=(12, 8), sharey=True)
axes = axes.flatten()

for i, (col, title, bins) in enumerate(plots):
    axes[i].hist(ssma[col].dropna(), bins=bins)
    axes[i].set_title(title, fontsize=11)
    axes[i].set_xlabel(title, fontsize=10)
    axes[i].set_ylabel("Count", fontsize=10)
    axes[i].grid(True, axis="y", alpha=0.3)

# Remove empty last subplot for a clean 2x3 layout
fig.delaxes(axes[-1])

fig.suptitle("Distributions of Key Numeric Variables", fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

Interpretation:
- Addicted_Score and Mental_Health_Score both show moderate spread indicating sufficient variability to analyze their relationship
- Average daily social media usage centers around 4–6 hours, suggesting many students engage heavily with social media.
- Sleep hours cluster around 6–8 hours, with some lower values that could relate to higher social media use or poorer mental health.
- Overall the distributions show reasonable variation across key variables, supporting further analysis of the relationship between social media addiction and mental health, and whether it differs by gender or academic level.

#### EDA: Academic-Level Stratified Analysis

This section compares key outcomes across `Academic_Level` (`High School`, `Undergraduate`, `Graduate`) to check whether usage, addiction score, and reported academic impact differ by level.


```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display

data_path = Path('data/00-raw/Students Social Media Addiction.csv')
if not data_path.exists():
    data_path = Path('Group029_WI26/data/00-raw/Students Social Media Addiction.csv')

ssma = pd.read_csv(data_path)

level_order = ['High School', 'Undergraduate', 'Graduate']
ssma['Academic_Level'] = pd.Categorical(ssma['Academic_Level'], categories=level_order, ordered=True)

summary = (
    ssma.groupby('Academic_Level', observed=False)
    .agg(
        n=('Student_ID', 'size'),
        median_usage_hrs=('Avg_Daily_Usage_Hours', 'median'),
        median_sleep_hrs=('Sleep_Hours_Per_Night', 'median'),
        median_addicted_score=('Addicted_Score', 'median'),
        performance_impact_rate=('Affects_Academic_Performance', lambda s: s.eq('Yes').mean())
    )
    .round(2)
)

summary_display = summary.copy()
summary_display['performance_impact_rate'] = (summary_display['performance_impact_rate'] * 100).round(1)
summary_display = summary_display.rename(columns={'performance_impact_rate': 'performance_impact_rate_%'})
display(summary_display)

sns.set_theme(style='whitegrid')
fig, axes = plt.subplots(1, 3, figsize=(16, 4.5))

sns.boxplot(data=ssma, x='Academic_Level', y='Avg_Daily_Usage_Hours', order=level_order, ax=axes[0])
axes[0].set_title('Daily Social Media Use by Academic Level')
axes[0].set_xlabel('Academic Level')
axes[0].set_ylabel('Hours per day')
axes[0].tick_params(axis='x', rotation=15)

sns.boxplot(data=ssma, x='Academic_Level', y='Addicted_Score', order=level_order, ax=axes[1])
axes[1].set_title('Addicted Score by Academic Level')
axes[1].set_xlabel('Academic Level')
axes[1].set_ylabel('Addicted Score')
axes[1].tick_params(axis='x', rotation=15)

perf_rate = (
    ssma.assign(perf_yes=ssma['Affects_Academic_Performance'].eq('Yes'))
    .groupby('Academic_Level', observed=False)['perf_yes']
    .mean()
    .reindex(level_order)
    .reset_index()
)

sns.barplot(data=perf_rate, x='Academic_Level', y='perf_yes', order=level_order, ax=axes[2], color='steelblue')
axes[2].set_title('Share Reporting Academic Impact')
axes[2].set_xlabel('Academic Level')
axes[2].set_ylabel('Proportion Yes')
axes[2].set_ylim(0, 1)
axes[2].tick_params(axis='x', rotation=15)

for i, row in perf_rate.iterrows():
    axes[2].text(i, row['perf_yes'] + 0.03, f"{row['perf_yes']:.0%}", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

**Interpretation of Data**

- Median daily usage and median addiction scores are somewhat higher in the High School group, while Undergraduate and Graduate patterns are closer to each other in this dataset.
- The percent reporting that social media affects academic performance is highest for High School students and lower for Undergraduate and Graduate students.
- The High School subgroup is much smaller (`n = 27`) than Undergraduate (`n = 353`) and Graduate (`n = 325`), so these differences should be treated as suggestive patterns rather than definitive conclusions.

#### EDA: Interaction exploration

This analysis examined whether the relationship between social media addiction and mental health differs by gender and academic level using interaction regression models.

##### Gender Interaction Model 

```python
import statsmodels.formula.api as smf

model_gender = smf.ols(
    'Mental_Health_Score ~ Addicted_Score * Gender',
    data=ssma
).fit()

print(model_gender.summary())
```

##### Gender Interaction Results 
The interaction between Addicted_Score and Gender was not statistically significant (β = -0.029, p = .094).

This indicates that:

The negative relationship between addiction and mental health is similar for males and females.

Although slopes appear slightly different visually, the difference is not statistically meaningful at the .05 level.

Conclusion:
Gender does not significantly moderate the relationship between social media addiction and mental health.

##### Academic Level Interaction

```python
model_academic = smf.ols(
    'Mental_Health_Score ~ Addicted_Score * Academic_Level',
    data=ssma
).fit()

print(model_academic.summary())
```

##### Academic Level Interaction Results 
The interaction between Addicted_Score and Academic_Level showed mixed results:

High School interaction term was significant (β = 0.221, p = .001)

Undergraduate interaction term was marginal (p = .053)

This suggests that:

The strength of the addiction–mental health relationship differs for high school students compared to the reference group.

The negative impact of addiction on mental health is weaker for high school students than for the reference academic level.

Conclusion:
Academic level partially moderates the relationship, particularly for high school students.

```python
import seaborn as sns

sns.lmplot(
    data=ssma,
    x='Addicted_Score',
    y='Mental_Health_Score',
    hue='Gender',
    ci=None
)
```

##### Overall Interpretation

Social media addiction strongly predicts poorer mental health.

This relationship does not differ by gender.

It does differ somewhat by academic level, especially for high school students.

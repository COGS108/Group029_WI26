## Background and Prior Work

Mental health concerns among college students have increased in recent years, with rising reports of anxiety and depression, coupled with the growth of social media platforms. Because most students engage with social networking sites daily, researchers have become interested in understanding whether online behavior plays a role in psychological well-being. While early discussions often focused on the amount of time spent online, more recent work has shifted toward examining the quality and consequences of social media use, including whether the behavior is compulsive, causes interpersonal conflict, or interferes with daily life. 


Social media use is extremely common among college students, but researchers now separate how much people use social platforms (time or frequency) from how they use them (such as compulsive or conflict-driven behavior). “Problematic social media use,” also described as social media addiction, refers to patterns that are hard to control and that disrupt daily life.<a name="cite_ref-1"></a><sup>1</sup> This distinction matters for our project because the dataset includes both an addiction-like score and relationship-focused variables, such as conflicts. This lets us examine whether problematic use is linked to worse mental health outcomes beyond what can be explained by time spent on social media alone.Do you like this personality?


Prior research consistently shows that higher levels of problematic social media use are linked to poorer mental health outcomes, including higher depression and anxiety symptoms and lower overall well-being.<a name="cite_ref-2"></a><sup>2</sup><a name="cite_ref-3"></a><sup>3</sup> However, many studies emphasize that causality is unclear and that these effects may depend on individual or social factors. One systematic review of adolescents and young adults found widespread links between problematic use and depression and stress, but noted that most studies were cross-sectional and therefore could not determine whether social media use leads to poorer mental health or vice versa.<a name="cite_ref-2"></a><sup>2</sup> Public health reports argue that risks cannot be captured by screen time alone and that the type of engagement and harmful experiences matter more than total time spent online.<a name="cite_ref-4"></a><sup>4</sup> These findings motivate two directions for our study: first, testing whether addiction-like measures of social media use are associated with mental health and whether that association varies by gender or academic level, and second, examining whether conflict-related experiences predict poorer mental health even after controlling for total usage time and relationship status.


Measurement research is also relevant to our analysis. Many studies assess problematic social media use using validated tools such as the Bergen Social Media Addiction Scale (BSMAS), which is based on core addiction features like salience, mood modification, tolerance, withdrawal, conflict, and relapse.<a name="cite_ref-1"></a><sup>1</sup> Validation studies show that the BSMAS has a stable structure and can reliably measure problematic use in young populations, making it useful for studying links with mental health and related outcomes.<a name="cite_ref-1"></a><sup>1</sup> Although our Kaggle dataset uses its own addiction score, this work helps clarify what this measure is intended to capture. 


Because one potential research direction focuses specifically on interpersonal conflict and relationship factors, earlier work on romantic and social relationships is especially relevant. Studies of college students and young adults have found that heavier social media use is associated with relationship strain.<a name="cite_ref-5"></a><sup>5</sup><a name="cite_ref-6"></a><sup>6</sup> In particular, Clayton et al. reported that greater Facebook use was linked to worse relationship outcomes especially in newer relationships.<a name="cite_ref-5"></a><sup>5</sup> Other research on “Facebook intrusion” reported similar connections between intrusive use, jealousy, and reduced satisfaction in undergraduate romantic relationships.<a name="cite_ref-6"></a><sup>6</sup> These results support our plan to examine whether conflicts over social media predict lower mental health scores, even after accounting for usage time and relationship status.


Finally, prior studies suggest that the relationship between problematic technology use and well-being can differ across demographic groups, which motivates our future analyses by gender and academic level. Research on smartphone addiction among undergraduates, for example, has reported gender differences in prevalence and associated factors, implying that risks and outcomes may not be the same for everyone.<a name="cite_ref-7"></a><sup>7</sup> Therefore, one possible research question for this study is whether the association between social media addiction scores and mental health differs by gender or academic level, rather than assuming a single overall effect for all students.

## Research Question

For this project, our interest is to answer the following question:

Is there, for students, a statistically significant relationship between social media addiction score and mental health score, and does this relationship differ by gender or academic level?

For context, we will define the social media addiction score and mental health score as the following:
* Social Media Addiction Score: a self-reported single-item measure on a 10-point Likert scale, where:
    *  1 = Very poor mental health (frequent distress, anxiety, low mood, difficulty coping)
    *  10 = Excellent mental health (emotionally stable, low stress, positive mood, good coping ability)
* Mental Health Score: a self-reported 10-point Likert scale measure, where:
    *  1 = No signs of addictive behavior (controlled, balanced use)
    *  10 = Severe addictive tendencies (compulsive use, inability to reduce usage, interference with sleep, academics, or relationships)

The respondents are mostly in the late teen to young adult range, roughly ages 16 to 25, so the dataset is best interpreted as a snapshot of student experiences rather than a general population sample.

There are a few major concerns to keep in mind. First, nearly all variables are self reported, so usage hours, sleep, mental health, and addiction severity can be affected by memory errors and by how comfortable someone is reporting honestly. Second, the data are cross sectional, meaning they capture one point in time, so you can study associations but you cannot say social media use causes changes in mental health, sleep, or relationship conflict. Third, the sample is likely shaped by who chose to respond, so it may not be representative of all students, and responses across countries can differ based on culture and how people interpret rating scales. Finally, some public writeups describe the score ranges slightly differently, so it is safer to verify the actual distributions directly in the dataset before using strict cutoffs or treating the scales as standardized clinical measures.

#### EDA: Competing Explanatory Variables

We will examine whether each candidate variable is associated with both addiction and mental health. We will then include the candidate variable in the base model and evaluate whether the coefficient of Addicted_Score changes meaningfully (≥10%). Variables that substantially alter the estimated effect will be considered potential confounders or competing explanatory variables.

Some potential candidates we are exploring are `Avg_Daily_Usage_Hours`, `Sleep_Hours_Per_Night`, and `Conflicts_Over_Social_Media`.

Let's first establish the baseline model. 

```python
relationship_mh_v_add = smf.ols(
    'Mental_Health_Score ~ Addicted_Score',
    data=ssma
).fit()
print(relationship_mh_v_add.summary())
baseline_model = relationship_mh_v_add
```

From earlier computation, we obtain the following about **Mental_Health_Score** vs. **Addicted_Score**:
1. Correlation: -0.945051
2. R-Squared: 0.893
3. Coefficient Size: -0.6580
4. P-Vlaue of the coefficient: 0.000

Let's first see how correlated the three variables are with `Addicted_Score` and `Mental_Health_Score`

```python
comp_variables = ['Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night', 'Conflicts_Over_Social_Media']

corr_data = {"comparison":['Addicted_Score'],
            "Correlation with Mental_Health_Score":[r],
            "Correlation with Addicted_Score":[1]}
corr_comp_table = pd.DataFrame(data=corr_data).set_index('comparison')

for comp in comp_variables:
    mh_v_comp_corr = ssma['Mental_Health_Score'].corr(ssma[comp])
    add_v_comp_corr = ssma['Addicted_Score'].corr(ssma[comp])

    corr_comp_table.loc[f'{comp}'] = [mh_v_comp_corr, add_v_comp_corr]

corr_comp_table
```

Here, we can see that in general all three variables are strongly correlated to both `Mental_Health_Score` and `Addicted_Score`. This also suggests the three variables have a strong linear association. Because potential confounders are typically associated with both the predictor and the outcome, these variables are reasonable candidates for confounding. Therefore, we will then extend the baseline model '`Mental_Health_Score` ~ `Addicted_Score`' by adding each variable to determine whether the coefficient of `Addicted_Score` changes substantially. A meaningful change in the coefficient would indicate that the variable may confound the relationship between Addicted_Score and Mental_Health_Score.

Here, all three variables are strongly correlated with both `Mental_Health_Score` and `Addicted_Score`, indicating a strong linear association. While these variables are related to both the predictor and the outcome, we cannot assume causality. To explore whether they influence the observed association, we extend the baseline model '`Mental_Health_Score` ~ `Addicted_Score`' by adding each variable individually. We then examine whether the coefficient of `Addicted_Score` changes substantially. A meaningful change in the coefficient would suggest that the variable may partially account for or explain the relationship between `Addicted_Score` and `Mental_Health_Score`.

```python
formula_comp_curr = {
    "formula": ['Addicted_Score'],
    "r_squared": [relationship_mh_v_add.rsquared],
    "add_coef": [relationship_mh_v_add.params.iloc[1]],
    "% change in add_coef": [0.0],
    "comp_coef": [relationship_mh_v_add.params.iloc[1]],
    "comp_pval": [relationship_mh_v_add.pvalues.iloc[1]]
}

model_comparison = pd.DataFrame(data = formula_comp_curr).set_index('formula')

comp_variables = ['Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night', 'Conflicts_Over_Social_Media']
for comp in comp_variables:
    formula = f'Mental_Health_Score ~ Addicted_Score + {comp}'
    relationship = smf.ols(formula, data=ssma).fit()
    rsquared = relationship.rsquared
    new_add_coef = relationship.params.iloc[1]
    change_in_add_coef = (new_add_coef - model_comparison.loc['Addicted_Score', 'add_coef'])/model_comparison.loc['Addicted_Score', 'add_coef']*100
    comp_coef = relationship.params.iloc[2]
    comp_pval = relationship.pvalues.iloc[2]
    
    model_comparison.loc[f'Addicted_Score + {comp}'] = [rsquared, new_add_coef, change_in_add_coef, comp_coef, comp_pval]
```

```python
model_comparison
```

By creating the above comparison table, we can see the following things:

1. `r_squared`: how much of the variance of y is explained by the formula
2. `add_coef`: the coefficient of the `Addicted_Score` after revision of formula
3. `% change in add_coef`: the percent change between the coefficient of the `Addicted_Score` in revised and the baseline.
4. `comp_coef`: the coefficient of the newly added variable.
5. `comp_pval`: the p-value of the  newly added variable.

For the addition of `Avg_daily_Usage_Hours`, compared to the baseline, we see that the `Addicted_Score` coefficient decreased 4.23% in magnitude. This suggests, `Avg_Daily_Usage_hours` does not substantially change the estimated effect of `Addicted_Score`.

For the addition of `Conflicts_Over_Social_Media`, compared to the baseline, we see that the `Addicted_Score` coefficient decreased 8.68% in magnitude. This suggests, `Conflicts_Over_Social_Media`does not substantially change the estimated effect of `Addicted_Score`.

If we observe the p_value of the new variables, we find most are statstically associated associated with `Mental_Health_Score`, with the exception of `Sleep` (greater than 0.05 by a sliver). However, they do not contribute as seen by the miniscule change in r_squared.

In conclusion, while we discovered strong linear association of the three variables with `Mental_Health_Score` and `Addicted_Score`, we discover no significant change in the coefficients of `Addicted_Score` after extending them individually to the baseline regression model. This suggests that these variables do not meaningful contribute to the association between `Mental_Health_Score` and `Addicted_Score` in this dataset.

#### Permutation Test: Addiction vs Mental Health

The permutation test showed a strong negative relationship between social media addiction and mental health scores. The observed correlation between addiction score and mental health score was r = -0.945. After generating 10,000 permutations of the mental health scores, none of the permuted correlations were as extreme as the observed correlation, resulting in a permutation p-value ≈ 0.000 (p < 0.001). This indicates that the observed correlation is highly unlikely to occur by chance. Therefore, the results suggest a statistically significant negative relationship between social media addiction and mental health, meaning that higher addiction scores are associated with lower mental health scores.

```python
obs_corr = ssma["Addicted_Score"].corr(ssma["Mental_Health_Score"])

n_perm = 10000
perm_corrs = []

for i in range(n_perm):
    shuffled = np.random.permutation(ssma["Mental_Health_Score"])
    perm_corr = np.corrcoef(ssma["Addicted_Score"], shuffled)[0,1]
    perm_corrs.append(perm_corr)

p_value = np.mean(np.abs(perm_corrs) >= abs(obs_corr))

print("Observed correlation:", obs_corr)
print("Permutation p-value:", p_value)
```

```python
plt.hist(perm_corrs, bins=40)

plt.axvline(obs_corr)

plt.title("Permutation Test: Social Media Addiction vs Mental Health")

plt.xlabel("Correlation between Addiction Score and Mental Health Score")
plt.ylabel("Number of Permutations")

plt.show()
```

#### Permutation Test: Does the Correlation Differ by Gender?

A permutation test was conducted to examine whether the relationship between social media addiction and mental health differs by gender. The observed difference in correlation between male and female students was -0.0025. After randomly permuting the gender labels 10,000 times, the permutation test produced a p-value of 0.7282. Because this p-value is much larger than 0.05, the observed difference in correlations is likely due to random variation. Therefore, the results suggest that the relationship between social media addiction and mental health does not significantly differ between male and female students.

```python
male = df[df["Gender"] == "Male"]
female = df[df["Gender"] == "Female"]

r_male = male["Addicted_Score"].corr(male["Mental_Health_Score"])
r_female = female["Addicted_Score"].corr(female["Mental_Health_Score"])

obs_diff = r_male - r_female
```

```python
perm_diffs = []

for i in range(10000):
    
    shuffled_gender = np.random.permutation(df["Gender"])
    df["perm_gender"] = shuffled_gender
    
    m = df[df["perm_gender"] == "Male"]
    f = df[df["perm_gender"] == "Female"]
    
    diff = m["Addicted_Score"].corr(m["Mental_Health_Score"]) - \
           f["Addicted_Score"].corr(f["Mental_Health_Score"])
    
    perm_diffs.append(diff)

p_value_gender = np.mean(np.abs(perm_diffs) >= abs(obs_diff))

print("Observed difference:", obs_diff)
print("Permutation p-value:", p_value_gender)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(perm_diffs, bins=40)

plt.axvline(obs_diff, linestyle='--')

plt.title("Permutation Test: Correlation Difference by Gender")
plt.xlabel("Permuted Difference in Correlation (Male − Female)")
plt.ylabel("Frequency")

plt.show()
```

#### Permutation Test — Does the Correlation Differ by Academic Level?

A permutation test was conducted to examine whether the relationship between social media addiction and mental health differs across academic levels. The observed difference between the highest and lowest correlations across academic groups was 0.0544. After randomly permuting the academic level labels 10,000 times, the permutation test produced a p-value of 0.0194. Since this p-value is below 0.05, the observed difference is unlikely to occur by chance. However, because the High School subgroup is substantially smaller than the undergraduate and graduate groups, its correlation estimate is less stable. Therefore, the observed difference should be interpreted with caution.

```python
groups = df["Academic_Level"].unique()

corrs = {}

for g in groups:
    sub = df[df["Academic_Level"] == g]
    corrs[g] = sub["Addicted_Score"].corr(sub["Mental_Health_Score"])

obs_stat = max(corrs.values()) - min(corrs.values())
```

```python
perm_stats = []

for i in range(10000):
    
    shuffled = np.random.permutation(df["Academic_Level"])
    df["perm_level"] = shuffled
    
    corrs = []
    
    for g in groups:
        sub = df[df["perm_level"] == g]
        corrs.append(sub["Addicted_Score"].corr(sub["Mental_Health_Score"]))
    
    stat = max(corrs) - min(corrs)
    perm_stats.append(stat)

p_value_level = np.mean(np.array(perm_stats) >= obs_stat)

print("Observed difference:", obs_stat)
print("Permutation p-value:", p_value_level)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(perm_stats, bins=40)

plt.axvline(obs_stat, linestyle='--')

plt.title("Permutation Test: Correlation Difference by Academic Level")
plt.xlabel("Permuted Difference in Correlations")
plt.ylabel("Frequency")

plt.show()
```

This finding is further supported by the permutation test, which confirms that the observed correlation is highly unlikely to occur by chance ($p<0.001$). Together, these results strongly support the first part of the research question: social media addiction and mental health are significantly related in this dataset.

However, the evidence for differences in this relationship across demographic groups is mixed. Nevertheless, because the High School subgroup is much smaller than the others, its correlation estimate is less stable and should be interpreted cautiously. Overall, while the primary hypothesis that addiction is negatively associated with mental health is strongly supported, the hypothesis that this relationship clearly differs by gender and academic level is only partially supported, with meaningful differences appearing mainly across academic levels rather than gender.

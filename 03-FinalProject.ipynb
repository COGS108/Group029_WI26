### Baseline Model: Mental Health Score vs. Addiction Score

The baseline association between `Addicted_Score` and `Mental_Health_Score` is very strong and negative. In this dataset, the correlation is `-0.945`, the model `R^2` is `0.893`, and the slope on `Addicted_Score` is `-0.658` with `p < 0.001`. This indicates that a one-point increase in social media addiction score is associated with about a `0.66` point decrease in mental health score on average, providing strong evidence that higher addiction levels are linked to worse mental health outcomes.

```python
import pandas as pd
import statsmodels.formula.api as smf
from pathlib import Path

if 'ssma' not in globals():
    for data_path in [
        Path('data/02-processed/Students Social Media Addiction_clean.csv'),
        Path('data/00-raw/Students Social Media Addiction.csv'),
    ]:
        if data_path.exists():
            ssma = pd.read_csv(data_path)
            break
    else:
        raise FileNotFoundError('Students Social Media Addiction dataset not found in data/00-raw or data/02-processed.')

baseline_model = smf.ols(
    'Mental_Health_Score ~ Addicted_Score',
    data=ssma
).fit()

baseline_results = pd.DataFrame({
    'Metric': ['Correlation', 'R-squared', 'Intercept', 'Addicted_Score coefficient', 'Addicted_Score p-value'],
    'Value': [
        ssma['Addicted_Score'].corr(ssma['Mental_Health_Score']),
        baseline_model.rsquared,
        baseline_model.params['Intercept'],
        baseline_model.params['Addicted_Score'],
        baseline_model.pvalues['Addicted_Score'],
    ],
})

baseline_results.round(4)
```

### Interaction Models: Testing Moderation by Gender and Academic Level

For gender, the interaction term is small and not statistically significant (`beta = -0.029`, `p = 0.094`), which suggests that the negative relationship between addiction score and mental health score is similar for male and female students. For academic level, the high school interaction term is significant (`beta = 0.221`, `p < 0.001`) and the undergraduate interaction is marginal (`p = 0.053`), indicating that academic level partially moderates this relationship. In particular, the association appears weaker for high school students relative to the graduate-student reference group.

```python
if 'ssma' not in globals():
    for data_path in [
        Path('data/02-processed/Students Social Media Addiction_clean.csv'),
        Path('data/00-raw/Students Social Media Addiction.csv'),
    ]:
        if data_path.exists():
            ssma = pd.read_csv(data_path)
            break
    else:
        raise FileNotFoundError('Students Social Media Addiction dataset not found in data/00-raw or data/02-processed.')

gender_interaction_model = smf.ols(
    'Mental_Health_Score ~ Addicted_Score * Gender',
    data=ssma
).fit()

academic_interaction_model = smf.ols(
    'Mental_Health_Score ~ Addicted_Score * Academic_Level',
    data=ssma
).fit()

interaction_results = pd.DataFrame([
    {
        'Model': 'Gender interaction',
        'Term': 'Addicted_Score:Gender[T.Male]',
        'Coefficient': gender_interaction_model.params['Addicted_Score:Gender[T.Male]'],
        'p-value': gender_interaction_model.pvalues['Addicted_Score:Gender[T.Male]'],
        'R-squared': gender_interaction_model.rsquared,
    },
    {
        'Model': 'Academic level interaction',
        'Term': 'Addicted_Score:Academic_Level[T.High School]',
        'Coefficient': academic_interaction_model.params['Addicted_Score:Academic_Level[T.High School]'],
        'p-value': academic_interaction_model.pvalues['Addicted_Score:Academic_Level[T.High School]'],
        'R-squared': academic_interaction_model.rsquared,
    },
    {
        'Model': 'Academic level interaction',
        'Term': 'Addicted_Score:Academic_Level[T.Undergraduate]',
        'Coefficient': academic_interaction_model.params['Addicted_Score:Academic_Level[T.Undergraduate]'],
        'p-value': academic_interaction_model.pvalues['Addicted_Score:Academic_Level[T.Undergraduate]'],
        'R-squared': academic_interaction_model.rsquared,
    },
])

interaction_results.round(4)
```

### Permutation Test: Addiction vs Mental Health

Addiction and Mental Health

The permutation test showed a strong negative relationship between social media addiction and mental health scores. The observed correlation between addiction score and mental health score was r = -0.945. After generating 10,000 permutations of the mental health scores, none of the permuted correlations were as extreme as the observed correlation, resulting in a permutation p-value ≈ 0.000 (p < 0.001). This indicates that the observed correlation is highly unlikely to occur by chance. Therefore, the results suggest a statistically significant negative relationship between social media addiction and mental health, meaning that higher addiction scores are associated with lower mental health scores.

```python
data_path = Path("data/00-raw/Students Social Media Addiction.csv")
df = pd.read_csv(data_path)

df.head()
```

### Permutation Test: Does the Correlation Differ by Gender?

Addiction and Mental Health by Gender

A permutation test was conducted to examine whether the relationship between social media addiction and mental health differs by gender. The observed difference in correlation between male and female students was -0.0025. After randomly permuting the gender labels 10,000 times, the permutation test produced a p-value of 0.7338. Because this p-value is much larger than 0.05, the observed difference in correlations is likely due to random variation. Therefore, the results suggest that the relationship between social media addiction and mental health does not significantly differ between male and female students.

```python
male = df[df["Gender"] == "Male"]
female = df[df["Gender"] == "Female"]

r_male = male["Addicted_Score"].corr(male["Mental_Health_Score"])
r_female = female["Addicted_Score"].corr(female["Mental_Health_Score"])

obs_diff = r_male - r_female
```

```python
perm_diffs = []

for i in range(10000):
    
    shuffled_gender = np.random.permutation(df["Gender"])
    df["perm_gender"] = shuffled_gender
    
    m = df[df["perm_gender"] == "Male"]
    f = df[df["perm_gender"] == "Female"]
    
    diff = m["Addicted_Score"].corr(m["Mental_Health_Score"]) - \
           f["Addicted_Score"].corr(f["Mental_Health_Score"])
    
    perm_diffs.append(diff)

p_value_gender = np.mean(np.abs(perm_diffs) >= abs(obs_diff))

print("Observed difference:", obs_diff)
print("Permutation p-value:", p_value_gender)
```

### Permutation Test — Does the Correlation Differ by Academic Level?

Addiction and Mental Health by Academic Level

A permutation test was conducted to examine whether the relationship between social media addiction and mental health differs across academic levels. The observed difference between the highest and lowest correlations across academic groups was 0.0544. After randomly permuting the academic level labels 10,000 times, the permutation test produced a p-value of 0.0205. Since this p-value is below 0.05, the observed difference is unlikely to occur by chance. These results suggest that the relationship between social media addiction and mental health significantly differs across academic levels, indicating that the strength of the correlation varies between different academic groups.

```python
groups = df["Academic_Level"].unique()

corrs = {}

for g in groups:
    sub = df[df["Academic_Level"] == g]
    corrs[g] = sub["Addicted_Score"].corr(sub["Mental_Health_Score"])

obs_stat = max(corrs.values()) - min(corrs.values())
```

```python
perm_stats = []

for i in range(10000):
    
    shuffled = np.random.permutation(df["Academic_Level"])
    df["perm_level"] = shuffled
    
    corrs = []
    
    for g in groups:
        sub = df[df["perm_level"] == g]
        corrs.append(sub["Addicted_Score"].corr(sub["Mental_Health_Score"]))
    
    stat = max(corrs) - min(corrs)
    perm_stats.append(stat)

p_value_level = np.mean(np.array(perm_stats) >= obs_stat)

print("Observed difference:", obs_stat)
print("Permutation p-value:", p_value_level)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(perm_stats, bins=40)

plt.axvline(obs_stat, linestyle='--')

plt.title("Permutation Test: Correlation Difference by Academic Level")
plt.xlabel("Permuted Difference in Correlations")
plt.ylabel("Frequency")

plt.show()
```
